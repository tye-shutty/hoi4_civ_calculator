```python
import hoi4_civ_calculator as calc

speed_mod_1x = [(1,1)]
speed_mod_2x = [(1,2)]
daily_reports = [{'civ': 15, 'mil': 0, 'ref': 0, 'doc': 0, 'goods': 0}]
inf = {'moselland': [1.7, 60, 0]}
con_queue1 = calc.make_queue([(('inf', 'moselland'),1),(('mil', 'moselland'),60)])
con_queue2 = calc.make_queue([(('mil', 'moselland'),60)])
final_day = 365*3
inf_mil_1x =calc.calculate(daily_reports, con_queue1, inf, final_day, speed_mod_1x)
inf_mil_2x =calc.calculate(daily_reports, con_queue1, inf, final_day, speed_mod_2x)
mil_1x =calc.calculate(daily_reports, con_queue2, inf, final_day, speed_mod_1x)
mil_2x =calc.calculate(daily_reports, con_queue2, inf, final_day, speed_mod_2x)
```

```python
print('It takes 12 factories to break even')
print('Because construction speed is multiplied with infrastructure, higher speed makes inf more effective')
calc.graph([inf_mil_1x,inf_mil_2x,mil_1x,mil_2x], ['mil'],
           ['1 inf then mil (1x con speed)','1 inf then mil (2x con speed)',
           'mil (1x con speed)','mil (2x con speed)'], [0.5,1])
```

```python
scenario_1 = [{'civ': 15, 'mil': 0, 'ref': 0, 'doc': 0, 'goods': 0}]
scenario_2 = [{'civ': 14, 'mil': 0, 'ref': 0, 'doc': 0, 'goods': 0}]
inf = {'moselland': [1.7, 100, 0],'rhineland': [1.7, 100, 0],}
con_queue1 = calc.make_queue([(('civ', 'moselland'),1),(('mil', 'moselland'),99),(('mil', 'rhineland'),100)])
con_queue2 = calc.make_queue([(('mil', 'moselland'),100),(('mil', 'rhineland'),100)])
final_day = 365*3
civ_mil_15civ =calc.calculate(scenario_1, con_queue1, inf, final_day, speed_mod_2x)
mil_15civ =calc.calculate(scenario_1, con_queue2, inf, final_day, speed_mod_2x)
civ_mil_14civ =calc.calculate(scenario_2, con_queue1, inf, final_day, speed_mod_2x)
mil_14civ =calc.calculate(scenario_2, con_queue2, inf, final_day, speed_mod_2x)
```

```python
calc.graph([civ_mil_14civ, mil_14civ, civ_mil_15civ, mil_15civ], ['mil'],
           ['1 civ then mil (start with 14 civ)','mil (start with 14 civ)',
           '1 civ then mil (start with 15 civ)','mil (start with 15 civ)'], [0.5,1])
print('increasing civ from 14 to 15 is far more effective than 15 to 16')
```

```python
def working_civ(report):
    for x in range(len(civ_mil)):
        report[x]['working_civ']=report[x]['civ']-report[x]['goods']
    return report
print('2x speed, 20% consumer goods')
calc.graph([working_civ(civ_mil), working_civ(mil)], ['mil','working_civ'],
          ['one line always dedicated to civ','mil only'], [0.5,1])
```
